```python
# requirement
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cv
import itertools

class Proximal_gradient_method:
    def __init__(self, lamda):
        self.w_1 = np.arange(-1.5,3,0.01)
        self.w_2 = np.arange(-1.5,3,0.02)
        self.W1, self.W2 = np.mgrid[-1.5:3:0.01, -1.5:3:0.02]
        # 
        self.Value = np.zeros((len(self.w_1), len(self.w_2)))
        # 
        self.A = np.array([[3, 0.5],[0.5, 1]])
        # 
        self.mu = np.array([[1],[2]])
        # 
        self.w_init = np.array([[3], [-1]])
        # 
        self.lamda = lamda
        
    def calc_eval_fun(self):
        # 
        for i in range(len(self.w_1)):
            for j in range(len(self.w_2)):
                w = np.vstack([self.w_1[i], self.w_2[j]])
                self.Value[i, j] = np.dot(np.dot((w - self.mu).T, self.A), w - self.mu) + self.lamda * (np.abs(self.w_1[i]) + np.abs(self.w_2[j]))
    
    def ops(self, mu, q):
        x_proj = np.zeros(mu.shape)
        for i in range(len(mu)):
            if mu[i] > q:
                x_proj[i] = mu[i] - q
            else:
                if np.abs(mu[i]) < q:
                    x_proj[i] = 0
                else:
                    x_proj[i] = mu[i] + q; 
                
        return x_proj
    
    def main(self):
        # cvx
        w_lasso     = cv.Variable((2, 1))
        J           = cv.quad_form(w_lasso - self.mu, self.A) +  self.lamda * cv.norm(w_lasso, 1)
        objective   = cv.Minimize(J)
        constraints = []
        prob        = cv.Problem(objective, constraints)
        result      = prob.solve(solver = cv.CVXOPT) 
        w_lasso     = w_lasso.value

        plt.contour(self.W1, self.W2, self.Value)
        
        L = 1.0 * np.max(np.linalg.eig(2 * self.A)[0])
        w_history = []
        wt        = self.w_init
        w_stack_1 = np.zeros(1001)
        w_stack_2 = np.zeros(1001)
        for t in range(1000):
            w_history.append(wt.T)
            # 
            w_stack_1[t] = wt[0, 0]
            w_stack_2[t] = wt[1, 0]
            # 
            grad = 2 * np.dot(self.A, wt - self.mu)
            wth  = wt - 1 / L * grad
            wt   = self.ops(wth, self.lamda * 1 / L)

        w_history = np.vstack(w_history)
        
        plt.figure(1)
        plt.plot(w_lasso[0], w_lasso[1], 'ko')
        plt.plot(w_history[:, 0], w_history[:, 1], 'ro-', markersize = 3, linewidth = 0.5)
        
        plt.figure(2)
        show_iter = 30
        if self.lamda == 2:
            plt.plot(np.abs(w_stack_1[:show_iter]), 'ro-', linewidth=0.5, markersize=1, label = 'lamda = 2')
        elif self.lamda == 4:
            plt.plot(np.abs(w_stack_1[:show_iter]), 'ro-', linewidth=0.5, markersize=1, label = 'lamda = 4')
        elif self.lamda == 6:
            plt.plot(np.abs(w_stack_1[:show_iter]), 'ro-', linewidth=0.5, markersize=1, label = 'lamda = 6')
            
        plt.legend()
        plt.yscale('log')
        plt.xlabel('iter')
        plt.ylabel('diff from the gold weight_1')
        
        plt.figure(3)
        show_iter = 30
        if self.lamda == 2:
            plt.plot(np.abs(w_stack_2[:show_iter]), 'ro-', linewidth=0.5, markersize=1, label = 'lamda = 2')
        elif self.lamda == 4:
            plt.plot(np.abs(w_stack_2[:show_iter]), 'ro-', linewidth=0.5, markersize=1, label = 'lamda = 4')
        elif self.lamda == 6:
            plt.plot(np.abs(w_stack_2[:show_iter]), 'ro-', linewidth=0.5, markersize=1, label = 'lamda = 6')
        plt.legend()
        plt.yscale('log')
        plt.xlabel('iter')
        plt.ylabel('diff from the gold weight_2')

pd = Proximal_gradient_method(2)
pd.calc_eval_fun()
pd.main()
```
